```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {}
with open("science_auths.json") as f:
    data = json.load(f)
```

```python
comments_gen = []
comments_sub = []
removed_gen = []
removed_sub = []
age = []

for key, val in data.items():
    if val and key != "ScienceModerator" and key !="AutoModerator":
        cur_comments_sub = val["num_sub"]
        #if cur_comments_sub <= 1 or np.random.randint(cur_comments_sub) == 1:
        cur_age = val["age"]/(60*60*24)
        cur_comments_gen = val["num_gen"]
        cur_removed_gen = val["num_gen_removed"]
        cur_removed_sub = val["num_sub_removed"]

        age.append(cur_age)
        comments_gen.append(cur_comments_gen)
        comments_sub.append(cur_comments_sub)
        removed_gen.append(cur_removed_gen)
        removed_sub.append(cur_removed_sub)
```

```python
plt.hist(age, bins=30) 
plt.ylabel('# Occurrences')
plt.xlabel('Account Age (days)');
```

```python
plt.hist(comments_gen, bins=30) 
plt.ylabel('# Occurrences')
plt.xlabel('Number of comments in the last month');
```

```python
comments_sub.sort()
comments_filtered = [comment for comment in comments_sub if comment < 20]
plt.hist(comments_filtered) 
plt.ylabel('# Occurrences')
plt.xlabel('Number of r/Science comments in the last month');
```

```python
plt.hist(removed_sub, bins=100) 
plt.ylabel('# Occurrences')
plt.xlabel('Number of r/Science removed comments in the last month');
```

```python
plt.hist(removed_gen, bins=100) 
plt.ylabel('# Occurrences')
plt.xlabel('Number of r/Science removed comments in the last month');
```
